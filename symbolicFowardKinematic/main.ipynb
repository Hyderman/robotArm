```python
from sympy import symbols, cos, sin
from sympy.matrices import Matrix
```

```python
q = symbols('q:6')
α = symbols('alpha:6')
a = symbols('a:6')
d = symbols('d:6')
```

```python

T = [None]*6
for i in range(6):
    T[i] = Matrix([[cos(q[i]), -sin(q[i]), 0, a[i]],
                   [sin(q[i]) * cos(α[i]), cos(q[i]) * cos(α[i]), -sin(α[i]), -sin(α[i]*d[i])],
                   [sin(q[i]) * sin(α[i]), cos(q[i]) * sin(α[i]), cos(α[i]), cos(α[i]*d[i])],
                   [0, 0, 0, 1]])
T
```

```python
Tglobal = 1
for Ti in T:
    Tglobal = Tglobal * Ti
Tglobal
```
